```python
from ultralytics import YOLO
import os

# Show dataset structure
for dirname, _, filenames in os.walk('/kaggle/input/version3'):
    print(dirname)
```

```python
model = YOLO("yolov8n.pt")  # load pre-trained YOLOv8 nano
```

```python
model.train(
    data="/kaggle/input/version3/data.yaml",
    imgsz=320,
    epochs=60,
    batch=4,
    name="elephant_yolov8n",
    patience=20,
    device=0
)
```

```python
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/elephant_yolov8n/weights/best.pt")
model.export(format="onnx", opset=12, simplify=True)
```

```python
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/elephant_yolov8n/weights/best.onnx", task="detect")
results = model.predict(
    source="/kaggle/input/version3/test/images",
    save=True,        # saves output images with boxes
    conf=0.5          # set confidence threshold
)
```

```python
metrics = model.val(data="/kaggle/input/version3/data.yaml")
```
